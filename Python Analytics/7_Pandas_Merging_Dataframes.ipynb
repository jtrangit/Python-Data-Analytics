```python
#importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset["train"].to_pandas()

#data cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
#filter df for jobs in the US
df_US = df[df["job_country"] == "United States"].copy()

#df_US
df_US["job_posted_month"] = df_US["job_posted_date"].dt.strftime("%B")
#df_US
df_US_pivot = df_US.pivot_table(index="job_posted_month", columns="job_title_short", aggfunc="size")
#the pivot table is showing the months in alphabetical order we need to put it in normal monthly order
df_US_pivot.reset_index(inplace=True)
df_US_pivot["month_no"] = pd.to_datetime(df_US_pivot["job_posted_month"], format="%B").dt.month

df_US_pivot["month_no"]
#df_US_pivot
#now we have the number associated with the months we can sort the months in order
df_US_pivot.sort_values("month_no", inplace=True)
df_US_pivot.set_index("job_posted_month", inplace=True)

#drop the month_no column now that we have it sorted 
df_US_pivot.drop(columns= "month_no", inplace=True)
df_US_pivot

```

```python
#We are going to merge the dataframe we were using last with a new external dataframe
df_US_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col="job_posted_month")
```

```python
#merge our dataframes, now we have a dataframe with data jobs and software jobs
df_US_merged = df_US_pivot.merge(df_US_software_pivot, on="job_posted_month")
```

```python
top_5 = (df_US_merged
         .sum()
         .sort_values(ascending=False)
         .head()
         .index
         .to_list())

top_5
```

```python
df_US_merged[top_5].plot(kind="line")
plt.title("Monthly Job Postings for Top Tech Jobs In the US")
plt.xlabel("2023")
plt.ylabel("Job Count")
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
